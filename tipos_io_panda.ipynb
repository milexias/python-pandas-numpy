###**Lendo arquivos JSON**

````python
import pandas as pd
````

````python
dados_pacientes = pd.read_json('/content/pacientes.json')
dados_pacientes.head()
````

````python
dados_pacientes_2 = pd.read_json('/content/pacientes_2.json')
dados_pacientes_2
````

###**Normalizando arquivos JSON**

````python
# deixando dados em formato tabular

df_normalizado = pd.json_normalize(dados_pacientes_2['Pacientes'])
df_normalizado
````

A função `json_normalize()` aceita apenas um dicionário ou uma lista de dicionários. Porém, se tentarmos usar parâmetros da função json_normalize em um arquivo JSON podem surgir erros. Para contornar isso, precisamos importar o módulo json e ler os arquivos conforme o código abaixo:

```
import pandas as pd
import json

# lendo o arquivo json usando o módulo Python JSON
with open('pacientes_2.json','r') as f:
    dados = json.loads(f.read())

# normalizando os dados com os parâmetros record_path e meta
pd.json_normalize(dados, record_path='Pacientes', meta=['Pesquisa', 'Ano'])

```





###**Escrevendo arquivos JSON**

````python
df_normalizado.to_json('historico_pacientes_norm.json')
pd.read_json('/content/historico_pacientes_norm.json') # pode salvar em outros formatos
````

Dados: https://github.com/vqrca/Pandas/blob/main/clientes_banco.csv

###**Criando um banco de dados local**

````python
import sqlalchemy
````

````python
pip install --upgrade 'sqlalchemy<2.0'  # necessário versão < 2.0 para rodar as funções
````

````python
sqlalchemy.__version__
````

````python
from sqlalchemy import create_engine, MetaData, Table, inspect

engine = create_engine('sqlite:///:memory:')
````

###**Escrevendo em um banco de dados**

````python
import pandas as pd

url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'

dados = pd.read_csv(url)
dados.head()
````

````python
dados.to_sql('clientes', engine, index=False)
````

````python
inspector = inspect(engine)

print(inspector.get_table_names())
````

###**Lendo uma consulta SQL**

````python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
````

````python
empregados = pd.read_sql(query, engine) # lendo uma consulta

empregados.head()
````

````python
empregados.to_sql('empregados2', con=engine, index=False) # engine conexão com o db
````

````python
pd.read_sql_table('empregados2', engine) # lendo a tabela inteira
````

````python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade',
                                                 'Rendimento_anual'])
````

###**Atualizando um banco de dados**
